## Introducción y Justificación del Tópico

La detección temprana y la clasificación precisa de las deficiencias cognitivas en individuos con síndrome de Down representan un reto crucial en el ámbito de la salud. El diagnóstico temprano permite orientar intervenciones específicas que favorezcan el desarrollo cognitivo, mejorando la calidad de vida y el pronóstico funcional de los pacientes. La información obtenida a través de pruebas de tamizaje prenatal, combinada con datos clínicos y demográficos, brinda una base cuantitativa que puede ser aprovechada mediante técnicas avanzadas de análisis de datos.

Esta investigación se apoya en fundamentos científicos y cuantitativos que integran conceptos de estadística, matemáticas y física aplicada, además de principios ingenieriles y medicos en la interpretación y transformación de datos. 

La aplicación de modelos de clasificación como la regresión logística, las máquinas de vectores de soporte y las redes neuronales, junto con la optimización de hiperparámetros mediante métodos bayesianos y otros metodos, permiten extraer insights precisos que complementen la evaluación clínica tradicional, la propuesta busca, a través de un análisis riguroso, identificar perfiles de riesgo y niveles de déficit cognitivo, lo que posibilitará diseñar intervenciones personalizadas para reducir estas deficiencias.

# Marco Teórico

Este marco teórico establece las bases matemáticas, estadísticas y computacionales que sustentan el desarrollo de modelos de clasificación para predecir el nivel de déficit cognitivo en individuos con síndrome de Down. Se abordan desde funciones de pérdida y teoría de espacios de Hilbert hasta la optimización de hiperparámetros y el análisis riguroso mediante métricas de evaluación. A continuación, se hace énfasis en la formalización matemática de cada concepto.

---

## 1. Función de Pérdida: Entropía Cruzada

La **entropía cruzada** es una medida de la discrepancia entre la distribución real $p$ y la distribución predicha $q$. Para un ejemplo en un problema binario, donde la variable objetivo $y \in \{0,1\}$ y la predicción es $\hat{y} \in (0,1)$, se define:

$$
L(y, \hat{y}) \;=\; -\Big[\,y \log(\hat{y}) \;+\; (1-y)\log\bigl(1-\hat{y}\bigr)\Big].
$$

Para clasificación multiclase (donde $y$ es un vector one-hot y $\hat{y}$ es un vector de probabilidades), la fórmula se generaliza a:

$$
L(y, \hat{y}) \;=\; -\sum_{i=1}^{K} y_i \log\bigl(\hat{y}_i\bigr).
$$

La función de pérdida de entropía cruzada se minimiza mediante algoritmos de optimización (por ejemplo, descenso del gradiente) para ajustar los parámetros del modelo, garantizando que la distribución de salida $\hat{y}$ se acerque a la distribución real $y$. Su interpretación probabilística la vincula con la **log-verosimilitud negativa**, ya que maximizar la verosimilitud de un modelo en problemas de clasificación es equivalente a minimizar la entropía cruzada.

**Referencia:** [Bishop (2006) – Pattern Recognition and Machine Learning](https://www.springer.com/gp/book/9780387310732)

---

## 2. Espacios de Hilbert y el Kernel Trick

Un **espacio de Hilbert** $\mathcal{H}$ es un espacio vectorial con un producto interno $\langle \cdot, \cdot \rangle$ que induce la norma:

$$
\| x \| \;=\; \sqrt{\langle x, x \rangle}.
$$

La completitud de $\mathcal{H}$ garantiza que toda sucesión de Cauchy converge en este espacio. En el contexto de *machine learning*, sobre todo en algoritmos como SVM y métodos de *kernel*, se explota el **kernel trick** para mapear los datos a un espacio de Hilbert de alta dimensión sin computar explícitamente la transformación $\phi(x)$. Se define un *kernel* $K$ que satisface:

$$
K(x, x') \;=\; \langle \phi(x), \phi(x') \rangle.
$$

Esto permite aplicar algoritmos lineales (por ejemplo, **SVM lineal**) en un espacio potencialmente no lineal respecto del original. El fundamento matemático del uso de kernels en espacios de Hilbert Reproducibles (RKHS) se basa en el Teorema de Representación de *Mercer*, que establece la correspondencia entre funciones *positivas definidas* y ciertos espacios de funciones donde el producto interno se define naturalmente a través del kernel.

**Referencia:** [Schölkopf & Smola (2002) – Learning with Kernels](https://mitpress.mit.edu/books/learning-kernels)

---

## 3. Regresión Logística

La **regresión logística** modela la probabilidad de que una instancia pertenezca a una clase mediante una combinación lineal de variables. Sea $x \in \mathbb{R}^n$ y $\mathbf{w} \in \mathbb{R}^n$, el modelo es:

$$
z \;=\; \beta_0 \;+\; \mathbf{w}^\top x.
$$

La salida se transforma usando la función sigmoide:

$$
\hat{y} \;=\; \sigma(z) \;=\; \frac{1}{1 + e^{-z}}.
$$

La función de pérdida que se emplea para ajustar $\mathbf{w}$ y $\beta_0$ es la entropía cruzada (o log-pérdida), equivalente a **maximizar** la verosimilitud bajo un modelo Bernoulli. Para controlar la complejidad del modelo y evitar *overfitting*, se añade **regularización**:

- **L2 (Ridge):**

  $$
  L_{\text{reg}}(\mathbf{w}) \;=\; L(\mathbf{w}) \;+\; \lambda \|\mathbf{w}\|^2.
  $$

- **L1 (Lasso):**

  $$
  L_{\text{reg}}(\mathbf{w}) \;=\; L(\mathbf{w}) \;+\; \lambda \|\mathbf{w}\|_1.
  $$

En la práctica, es frecuente usar optimizadores como *descenso del gradiente estocástico* (SGD) o *métodos de segundo orden* (Newton-Raphson) para estimar los parámetros.

**Referencia:** [Hastie, Tibshirani & Friedman (2009) – The Elements of Statistical Learning](https://web.stanford.edu/~hastie/ElemStatLearn/)

---

## 4. Máquinas de Vectores de Soporte (SVM y SVC)

Las **SVM** buscan el hiperplano que maximiza el margen entre dos clases, asumiendo un problema binario donde $y \in \{-1, +1\}$. El problema de optimización en su forma lineal se expresa como:

$$
\min_{\mathbf{w}, b}\; \frac{1}{2}\|\mathbf{w}\|^2 
\quad 
\text{sujeto a} 
\quad 
y^{(i)}\bigl(\mathbf{w}^\top x^{(i)} + b\bigr) \;\ge\; 1, \quad \forall i.
$$

Para datos no separables linealmente, se introducen variables de holgura $\xi_i$ y el parámetro $C$:

$$
\min_{\mathbf{w}, b, \{\xi_i\}} \;\; 
\frac{1}{2}\|\mathbf{w}\|^2 \;+\; 
C \sum_{i=1}^{N} \xi_i
\quad
\text{sujeto a}
\quad
y^{(i)}(\mathbf{w}^\top x^{(i)} + b) \;\ge\; 1 \;-\; \xi_i, 
\quad 
\xi_i \;\ge\; 0.
$$

La extensión para clasificación multiclase (con estrategias como **One-vs-Rest** o **One-vs-One**) se denomina **SVC** en diversas librerías.

### 4.1 Tipos de Kernels

El poder de las SVM se incrementa usando funciones kernel que capturan relaciones más complejas en el espacio de atributos:

- **Kernel RBF (Radial Basis Function):**

  $$
  K(x, x') \;=\; \exp\bigl(-\gamma \,\|x - x'\|^2\bigr).
  $$

- **Kernel Sigmoid:**

  $$
  K(x, x') \;=\; \tanh\bigl(\alpha \, x^\top x' \;+\; c\bigr).
  $$

- **Kernel Polinómico:**

  $$
  K(x, x') \;=\; \bigl(\gamma \, x^\top x' \;+\; r\bigr)^d.
  $$

De manera formal, un kernel debe ser **positivo semidefinido** para garantizar la existencia de un espacio de Hilbert asociado (RKHS) donde la separación (o regresión) tenga solución lineal.

**Referencia:** [Cortes & Vapnik (1995) – Support-Vector Networks](https://link.springer.com/article/10.1007/BF00994018)

## 5. Redes Neuronales Multicapa (Multi-layer Perceptron - MLP)

El **MLP** es un modelo de redes neuronales *feedforward* compuesto por capas de neuronas que se interconectan secuencialmente:

- **Capa de Entrada:** Representa el vector $x$ (características).
- **Capas Ocultas:** Cada neurona en una capa oculta calcula:
  $$
  a_j \;=\; \phi\bigl(\sum_{i=1}^{n} w_{ij} x_i \;+\; b_j\bigr),
  $$
  donde $\phi$ es una función de activación (no lineal).
- **Capa de Salida:** Transforma la salida de la última capa oculta en probabilidades o valores deseados. Para clasificación multiclase se usa **softmax**:
  $$
  \hat{y}_j \;=\; \frac{e^{z_j}}{\sum_{k=1}^{K} e^{z_k}}.
  $$

### Funciones de Activación

- **ReLU (Rectified Linear Unit):**
  $$
  \phi(z) \;=\; \max(0,\, z).
  $$
- **Sigmoid:**
  $$
  \phi(z) \;=\; \frac{1}{1 + e^{-z}}.
  $$
- **Tanh:**
  $$
  \phi(z) \;=\; \tanh(z).
  $$

La **optimización** de los parámetros (pesos y sesgos) se realiza con algoritmos de retropropagación del error (backpropagation) combinados con optimizadores como *Adam* o *SGD*. El objetivo es minimizar una función de pérdida, típicamente la entropía cruzada en tareas de clasificación.

**Referencia:** [Goodfellow, Bengio & Courville (2016) – Deep Learning](https://www.deeplearningbook.org/)

# 6. Métricas de Evaluación para Clasificación

Las métricas cuantifican el rendimiento y la capacidad de generalización de los modelos. 
---

### Matriz de Confusión

La matriz de confusión es una herramienta fundamental que permite desglosar el comportamiento de un clasificador en términos de:
- **TP (Verdaderos Positivos):** Número de casos positivos correctamente clasificados.
- **FP (Falsos Positivos):** Número de casos negativos clasificados incorrectamente como positivos.
- **TN (Verdaderos Negativos):** Número de casos negativos correctamente clasificados.
- **FN (Falsos Negativos):** Número de casos positivos clasificados incorrectamente como negativos.

En formato de tabla, se representa de la siguiente manera:

|                   | Predicción Positiva | Predicción Negativa |
|-------------------|---------------------|---------------------|
| **Real Positiva** | TP                  | FN                  |
| **Real Negativa** | FP                  | TN                  |

---

### Accuracy

El *accuracy* (precisión global) mide la proporción de predicciones correctas sobre el total de predicciones. Matemáticamente se define como:

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}.
$$

---

### Precision (Precisión)

La *precision* indica la proporción de predicciones positivas correctas respecto a todas las predicciones positivas:

$$
\text{Precision} = \frac{TP}{TP + FP}.
$$

---

### Recall (Sensibilidad)

El *recall* mide la capacidad del modelo para identificar correctamente las instancias positivas:

$$
\text{Recall} = \frac{TP}{TP + FN}.
$$

---

### F1-Score

El *F1-Score* es la media armónica entre precision y recall, y se define como:

$$
\text{F1} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}.
$$

Existen variantes de F1-Score, como:
- **f1\_micro:** Calcula la F1 considerando la suma total de TP, FP y FN.
- **f1\_macro:** Calcula la F1 por cada clase y luego toma la media aritmética, sin ponderar por el número de ejemplos.
- **f1\_weighted:** Pondera la F1 de cada clase de acuerdo con el número de ejemplos de cada clase.
- **f1\_samples:** Calcula la F1 para cada muestra y luego promedia los resultados (usado en problemas multilabel).

---

### ROC AUC

El *ROC AUC* (Área Bajo la Curva ROC) cuantifica la capacidad del modelo para discriminar entre clases a distintos umbrales. La curva ROC se construye graficando la *tasa de verdaderos positivos* ($TPR$ o *Recall*) contra la *tasa de falsos positivos* ($FPR$):

$$
FPR = \frac{FP}{FP + TN}.
$$

El *AUC* es el área bajo esta curva, y un valor cercano a 1 indica un modelo con excelente capacidad discriminativa.

---

### Tasas de Falsos Positivos y Falsos Negativos

- **Tasa de Falsos Positivos (FPR):**

$$
\text{FPR} = \frac{FP}{FP + TN}.
$$

- **Tasa de Falsos Negativos (FNR):**

$$
\text{FNR} = \frac{FN}{TP + FN}.
$$

---

### Balanced Accuracy

La *balanced accuracy* es la media de la sensibilidad (recall) obtenida para cada clase, y es especialmente útil cuando se tienen clases desbalanceadas:

$$
\text{Balanced Accuracy} = \frac{1}{K}\sum_{i=1}^{K} \text{Recall}_i.
$$

---

### Otras Métricas

Existen otras métricas que permiten evaluar distintos aspectos del rendimiento del modelo:
- **top\_k\_accuracy:** Evalúa si la clase verdadera se encuentra entre las $k$ predicciones con mayor probabilidad.
- **average\_precision:** Promedia la precisión en diferentes umbrales, dando un resumen del rendimiento.
- **neg\_brier\_score:** Es el negativo del Brier score, que mide el error cuadrático medio entre las probabilidades predichas y las etiquetas reales.
- **neg\_log\_loss:** Penaliza fuertemente las predicciones erróneas que asignan baja probabilidad a la clase correcta.
- **jaccard:** Mide la similitud entre conjuntos, útil en problemas multilabel.
- **roc\_auc\_ovr, roc\_auc\_ovo, roc\_auc\_ovr\_wei:** Son variantes del ROC AUC utilizadas en clasificación multiclase, mediante estrategias One-vs-Rest, One-vs-One y ponderaciones.

**Referencia:**  
- Pedregosa, F., et al. (2011). *Scikit-learn: Machine Learning in Python*. Journal of Machine Learning Research, 12, 2825–2830. [Link a JMLR](http://jmlr.org/papers/v12/pedregosa11a.html)

# 7. Optimización de Hiperparámetros

La optimización de hiperparámetros es fundamental para mejorar la capacidad de generalización de un modelo y evitar tanto el *overfitting* como el *underfitting*. La idea es encontrar el conjunto óptimo de hiperparámetros $\theta$ que minimicen la función de pérdida $L(\theta)$.

$$
\theta^* = \arg\min_{\theta \in \Theta} L(\theta)
$$

### Métodos de Búsqueda de Hiperparámetros

#### 7.1 Grid Search
Grid Search explora exhaustivamente un espacio de hiperparámetros definido. Si $\Theta_1, \Theta_2, \dots, \Theta_m$ son los conjuntos de valores para cada hiperparámetro, el espacio de búsqueda se define como:

$$
\Theta = \Theta_1 \times \Theta_2 \times \cdots \times \Theta_m.
$$

Grid Search evalúa todas las combinaciones posibles para encontrar la que minimice $L(\theta)$.

#### 7.2 Random Search
Random Search selecciona aleatoriamente combinaciones de hiperparámetros dentro del espacio $\Theta$. Esto puede ser más eficiente en espacios de alta dimensión, ya que reduce el costo computacional en comparación con la búsqueda exhaustiva.

#### 7.3 Algoritmos Evolutivos
Los algoritmos evolutivos emplean operadores inspirados en la biología, como la selección, cruce y mutación, para explorar globalmente el espacio de hiperparámetros $\Theta$. Este enfoque puede ayudar a escapar de óptimos locales y encontrar soluciones robustas.

#### 7.4 Optimización Bayesiana
La optimización bayesiana utiliza un modelo probabilístico para aproximar la función de pérdida $f(\theta)$ en función de los hiperparámetros $\theta$. Típicamente, se utiliza un **Gaussian Process** para modelar $f(\theta)$ y se define una función de adquisición $a(\theta)$ (por ejemplo, *Expected Improvement*). La estrategia consiste en elegir:

$$
\theta^* = \arg\max_{\theta \in \Theta} \; a(\theta)
$$

Este método balancea la exploración (probar nuevas regiones del espacio) y la explotación (refinar soluciones prometedoras), lo que permite ajustar los hiperparámetros de manera más eficiente.

**Referencia:**  
- Zhang, H. (2018). *Bayesian Optimization for Hyperparameter Tuning*. [Link a arXiv](https://arxiv.org/abs/1807.02811)

## 8. Preprocesamiento y Transformación de Datos

El éxito del modelado depende del adecuado preprocesamiento y transformación de los datos para eliminar sesgos, escalar los valores y codificar variables de modo correcto.

### 8.1 Codificación de Variables Categóricas

- **One-Hot Encoding:**  
  Para una variable categórica $x$ con $K$ categorías, se transforma cada muestra en un vector $v \in \{0,1\}^K$ donde:
  
  $$
  v_i \;=\;
  \begin{cases}
  1 & \text{si } x \text{ pertenece a la categoría } i,\\
  0 & \text{en otro caso}.
  \end{cases}
  $$
  
- **Get Dummies:**  
  Implementación práctica de *one-hot encoding*, disponible en varias bibliotecas de análisis de datos.

### 8.2 Normalización y Estandarización

- **Normalización (Min-Max Scaling):**  
  Escala cada valor $x$ al rango $[0,1]$:
  $$
  x' \;=\; \frac{x - \min(x)}{\max(x) - \min(x)}.
  $$

- **Estandarización (Z-Score):**  
  Transforma $x$ para que tenga media cero y varianza uno:
  $$
  x' \;=\; \frac{x - \mu}{\sigma}.
  $$

### 8.3 Representación de Datos y Tensores

En tareas complejas (como redes neuronales), los datos se organizan en **tensores**, arreglos multidimensionales denotados:

$$
T \;\in\; \mathbb{R}^{n_1 \times n_2 \times \cdots \times n_d}.
$$

Esta notación facilita la aplicación de operaciones vectoriales y matriciales de forma eficiente.

## 9. Conceptos de Ajuste del Modelo


### 9.1 Overfitting y Underfitting

- **Overfitting:**  
  El modelo se ajusta demasiado a los datos de entrenamiento, aprendiendo ruido o patrones espurios. El error de entrenamiento es muy bajo, pero el error de validación o de prueba (generalización) es alto.

- **Underfitting:**  
  El modelo es demasiado simple para capturar la complejidad de la función verdadera $f(x)$, resultando en error alto tanto en entrenamiento como en validación.

### 9.2 Sesgo y Varianza

- **Sesgo (Bias):**  
  Error sistemático o diferencia entre la función verdadera $f(x)$ y la estimación promedio $\mathbb{E}[\hat{f}(x)]$. Formalmente:
  $$
  \text{Bias} \;=\; \mathbb{E}[\hat{f}(x)] \;-\; f(x).
  $$

- **Varianza:**  
  Cuantifica la sensibilidad del modelo a variaciones en la muestra de entrenamiento. Se define como:
  $$
  \text{Var}\bigl(\hat{f}(x)\bigr) 
  \;=\; 
  \mathbb{E}\Bigl[\bigl(\hat{f}(x) - \mathbb{E}[\hat{f}(x)]\bigr)^2\Bigr].
  $$

Existe un compromiso inherente (trade-off) entre sesgo y varianza, esencial para lograr buena capacidad de generalización.

### 9.3 Data Leakage

El **data leakage** ocurre cuando información del conjunto de prueba o de validación se introduce indebidamente durante el entrenamiento, inflando el rendimiento de manera artificial. Para evitarlo, es crucial separar estrictamente los datos de entrenamiento, validación y prueba, además de realizar transformaciones (p. ej. escalado) **después** de dicha partición.

# Marco Teórico Interdisciplinario para la Detección Temprana y Clasificación de Deficiencias Cognitivas en el Síndrome de Down

## I. Introducción: El Reto Interdisciplinario de las Deficiencias Cognitivas en el Síndrome de Down

El síndrome de Down (SD), la causa genética más común de discapacidad intelectual, afecta aproximadamente a 1 de cada 700 nacidos vivos.¹ Una característica constante en las personas con SD es la presencia de retrasos cognitivos, que varían de leves a moderados.² La identificación temprana de estas deficiencias es crucial, ya que permite implementar intervenciones específicas que favorecen el desarrollo cognitivo, mejorando la calidad de vida y el pronóstico funcional.

El reto que presenta el SD se fundamenta en la interacción multifactorial de variables genéticas, neurobiológicas, neuroquímicas y ambientales. Esta complejidad demanda un enfoque interdisciplinario en el que se integren conceptos de matemáticas formales, física, biología, química, medicina, biotecnología, neurociencia, ingeniería genética y biomedicina para explicar y modelar los mecanismos subyacentes.

## II. La Arquitectura Genética del Síndrome de Down y sus Consecuencias Neurológicas

### II.A. Descripción Formal de la Trisomía 21

El SD es causado por la presencia de una copia adicional del cromosoma 21 ($\mathrm{Hsa21}$). Se presenta en tres formas principales:

- **Trisomía 21:** Aproximadamente el 95% de los casos, debido a la *no disyunción* en la meiosis.
- **Síndrome de Down por Translocación:** Se une una parte del cromosoma 21 a otro cromosoma.
- **Síndrome de Down Mosaico:** Solo un subconjunto de células contiene la copia adicional.

El efecto fenotípico se explica mediante el aumento en la **dosis génica**. Sea $( E(g_i) $) la expresión del gen $( g_i $) en condiciones normales, la presencia de una copia extra implica que la expresión se puede modelar como:

$$
E'(g_i) = E(g_i) \times \left(1 + \frac{1}{2}\right) = \frac{3}{2}E(g_i),
$$

lo que altera la regulación y la proporción de proteínas que intervienen en procesos celulares críticos.

### II.B. Mecanismos Moleculares y Celulares del Desequilibrio Cromosómico

La sobreexpresión de genes localizados en el cromosoma 21 afecta el desarrollo del sistema nervioso central (SNC) mediante mecanismos cuantificables. Sea $( \vec{X} = (x_1, x_2, \dots, x_n) $) el vector de concentraciones de productos génicos, el incremento en la dosis génica se traduce en un escalamiento del vector:

$$
\vec{X}' = \frac{3}{2} \vec{X}.
$$

Este escalamiento perturba la cinética de reacciones bioquímicas en las que participan enzimas y receptores, modeladas mediante ecuaciones de velocidad de tipo Michaelis-Menten:

$$
v = \frac{V_{\max}[S]}{K_m + [S]},
$$

donde la concentración de sustrato \([S]\) puede alterarse por la sobreexpresión génica, modificando la tasa \(v\) de las reacciones celulares y, en consecuencia, afectando la señalización celular, por ejemplo, en las vías *WNT* y *Notch*.

# Vías de Señalización WNT y Notch

## **1. Vía WNT**
La vía WNT regula procesos clave como la proliferación celular, diferenciación y desarrollo embrionario. En su mecanismo canónico, la β-catenina juega un papel crucial.

### **Formalización matemática de la acumulación de β-catenina**
La acumulación de la β-catenina en el núcleo puede modelarse mediante una ecuación diferencial que balancea su producción ($P(t)$) y degradación ($D(t)$):

$$
\frac{d[\beta\text{-catenina}]}{dt} = P(t) - D(t)
$$

Donde:
- $P(t)$ es proporcional a la activación de la vía WNT.
- $D(t)$ puede describirse como $k_d \cdot [\beta\text{-catenina}]$, donde $k_d$ es la constante de degradación.

### **Expresión de genes dependientes de WNT**
El nivel de expresión génica inducido por β-catenina ($E_g$) puede modelarse como:

$$
E_g = \frac{E_{\text{max}} \cdot [\beta\text{-catenina}]}{K_m + [\beta\text{-catenina}]}
$$

Donde $E_{\text{max}}$ es la expresión máxima y $K_m$ es la constante de afinidad.

**Referencia**: Sokol, S. Y. (2011). *Wnt Signaling in Embryonic Development*. Elsevier.

---
## **2. Vía Notch**
La vía Notch es esencial en la comunicación célula-célula y controla decisiones celulares como diferenciación y apoptosis.

### **Formalización del procesamiento del receptor Notch**
Cuando la proteína Notch interactúa con su ligando, se activa una cascada de señalización intracelular. La cantidad de receptores Notch activados ($N_a$) puede modelarse como una función sigmoide respecto a la concentración del ligando ($[L]$):

$$
N_a = \frac{N_{\text{max}} \cdot [L]}{K_m + [L]}
$$

Donde:
- $N_{\text{max}}$ es la cantidad máxima de receptores activados.
- $K_m$ representa la afinidad del ligando-receptor.

### **Regulación génica**
El efecto de Notch en la transcripción génica puede representarse como:

$$
\frac{d[G]}{dt} = k_a \cdot N_a - k_d \cdot [G]
$$

Donde:
- $[G]$ es el nivel del gen regulado.
- $k_a$ y $k_d$ son las tasas de activación y degradación, respectivamente.

**Referencia**: Reichrath, J., & Reichrath, S. (2020). *Notch Signaling in Embryology and Cancer*. Springer.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import odeint
from scipy.fft import fft, fftfreq
from sklearn.decomposition import PCA
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import odeint

# Configuración de estilo con seaborn
sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (8, 5)

##############################################################
# 1. EXPRESIÓN GÉNICA: Normal vs Síndrome de Down
##############################################################

def plot_gene_expression_comparative(E=1.0, factor=0.5):
    """
    Se compara la expresión génica normal, E(g_i), vs la expresión en SD,
    donde E'(g_i) = (1 + factor) * E(g_i).
    Se usa fill_between para resaltar el incremento.
    """
    # Se generan valores constantes para graficar como línea horizontal
    x = np.array([0, 1])
    normal = np.array([E, E])
    sd = normal * (1 + factor)
    
    fig, ax = plt.subplots()
    ax.plot(x, normal, 'o-', color='skyblue', label='Normal: E(g_i)')
    ax.plot(x, sd, 'o-', color='salmon', label="SD: E'(g_i)")
    # Rellenar el área entre las dos líneas
    ax.fill_between(x, normal, sd, color='gray', alpha=0.3, label='Incremento 50%')
    ax.set_xticks([0,1])
    ax.set_xticklabels(['Normal', 'SD'])
    ax.set_ylabel('Nivel de expresión')
    ax.set_title('Comparativa de Expresión Génica')
    ax.legend()
    plt.show()

plot_gene_expression_comparative()

##############################################################
# 2. CINÉTICA ENZIMÁTICA: Michaelis-Menten
##############################################################

def plot_michaelis_menten_comparative(Vmax_normal=10, Km_normal=2, factor=0.5):
    """
    Se comparan las curvas de Michaelis-Menten para condiciones normales y SD.
    En SD se asume que Vmax se incrementa en un 50%.
    """
    S = np.linspace(0, 20, 300)
    v_normal = (Vmax_normal * S) / (Km_normal + S)
    Vmax_sd = Vmax_normal * (1 + factor)
    v_sd = (Vmax_sd * S) / (Km_normal + S)
    
    fig, ax = plt.subplots()
    ax.plot(S, v_normal, 'b-', label=f'Normal: Vmax={Vmax_normal}, Km={Km_normal}')
    ax.plot(S, v_sd, 'r-', label=f'SD: Vmax={Vmax_sd:.1f}, Km={Km_normal}')
    ax.fill_between(S, v_normal, v_sd, color='gray', alpha=0.3)
    ax.set_xlabel('Concentración [S]')
    ax.set_ylabel('Velocidad v')
    ax.set_title('Comparativa de Cinética Michaelis-Menten')
    ax.legend()
    plt.show()

plot_michaelis_menten_comparative()

##############################################################
# 3. VÍA WNT: Acumulación de β‑catenina
##############################################################

def beta_catenin_model(y, t, P, k_d):
    return P - k_d * y

def plot_beta_catenin_comparative(P_normal=1.0, k_d=0.2, factor=0.5):
    """
    Se comparan las curvas de acumulación de β-catenina para condiciones normales y SD.
    En SD se asume que la tasa de producción P aumenta en un 50%.
    Se utiliza fill_between para resaltar el área entre curvas.
    """
    t = np.linspace(0, 50, 500)
    sol_normal = odeint(beta_catenin_model, 0, t, args=(P_normal, k_d)).flatten()
    P_sd = P_normal * (1 + factor)
    sol_sd = odeint(beta_catenin_model, 0, t, args=(P_sd, k_d)).flatten()
    
    fig, ax = plt.subplots()
    ax.plot(t, sol_normal, 'b-', label=f'Normal: P={P_normal}')
    ax.plot(t, sol_sd, 'r-', label=f'SD: P={P_sd:.1f}')
    ax.fill_between(t, sol_normal, sol_sd, color='gray', alpha=0.3)
    ax.set_xlabel('Tiempo')
    ax.set_ylabel(r'Concentración $[\beta\mathrm{-catenina}]$')
    ax.set_title('Comparativa de Acumulación de β-catenina (Vía WNT)')
    ax.legend()
    plt.show()

plot_beta_catenin_comparative()

##############################################################
# 4. VÍA NOTCH: Activación y Regulación Génica
##############################################################

def notch_activation(L, N_max, Km):
    return (N_max * L) / (Km + L)

def plot_notch_activation_comparative(N_max_normal=1.0, Km_normal=1.0, factor=0.5):
    """
    Se comparan las curvas de activación de la vía Notch para condiciones normales y SD.
    Se asume que en SD N_max aumenta en un 50%.
    """
    L = np.linspace(0, 10, 200)
    N_a_normal = notch_activation(L, N_max_normal, Km_normal)
    N_max_sd = N_max_normal * (1 + factor)
    N_a_sd = notch_activation(L, N_max_sd, Km_normal)
    
    fig, ax = plt.subplots()
    ax.plot(L, N_a_normal, 'b-', label=f'Normal: N_max={N_max_normal}')
    ax.plot(L, N_a_sd, 'r-', label=f'SD: N_max={N_max_sd:.1f}')
    ax.fill_between(L, N_a_normal, N_a_sd, color='gray', alpha=0.3)
    ax.set_xlabel('Concentración del ligando [L]')
    ax.set_ylabel('Receptores activados')
    ax.set_title('Comparativa de Activación Notch')
    ax.legend()
    plt.show()

plot_notch_activation_comparative()

def gene_regulation_model(G, t, k_a, N_a, k_d):
    return k_a * N_a - k_d * G

def plot_gene_regulation_comparative(k_a=0.5, N_a_normal=1.0, k_d=0.1, factor=0.5):
    """
    Se compara la evolución de la expresión génica para condiciones normales y SD.
    Se asume que en SD N_a aumenta en un 50% (p. ej. mayor activación por Notch).
    """
    t = np.linspace(0, 50, 500)
    sol_normal = odeint(gene_regulation_model, 0, t, args=(k_a, N_a_normal, k_d)).flatten()
    N_a_sd = N_a_normal * (1 + factor)
    sol_sd = odeint(gene_regulation_model, 0, t, args=(k_a, N_a_sd, k_d)).flatten()
    
    fig, ax = plt.subplots()
    ax.plot(t, sol_normal, 'b-', label=f'Normal: N_a={N_a_normal}')
    ax.plot(t, sol_sd, 'r-', label=f'SD: N_a={N_a_sd:.1f}')
    ax.fill_between(t, sol_normal, sol_sd, color='gray', alpha=0.3)
    ax.set_xlabel('Tiempo')
    ax.set_ylabel('Expresión génica [G]')
    ax.set_title('Comparativa de Regulación Génica (Vía Notch)')
    ax.legend()
    plt.show()

plot_gene_regulation_comparative()
```

### II.C. Identificación de Genes Clave y sus Implicaciones

Entre los genes de interés se encuentran:

- **APP (Proteína Precursora Amiloide):**  
  La sobreexpresión de APP conduce a una mayor producción de fragmentos peptídicos \(\beta\)-amiloides (\(\mathrm{A}\beta\)), cuya acumulación puede modelarse por la ecuación de balance:

  $$
  \frac{d[\mathrm{A}\beta]}{dt} = k_{\text{prod}} \cdot E'(\text{APP}) - k_{\text{deg}} \cdot [\mathrm{A}\beta],
  $$

  donde \(k_{\text{prod}}\) y \(k_{\text{deg}}\) son las constantes de producción y degradación, respectivamente.

- **SOD1 (Superóxido Dismutasa 1):**  
  La sobreexpresión conduce a un exceso en la conversión del anión superóxido \((\mathrm{O_2^-})\) a peróxido de hidrógeno (\(\mathrm{H_2O_2}\)). La reacción puede representarse como:

  $$
  2\,\mathrm{O_2^-} + 2\,\mathrm{H^+} \xrightarrow{\text{SOD1}} \mathrm{H_2O_2} + \mathrm{O_2}.
  $$

- **S100B:**  
  Este gen codifica una proteína con función neurotrófica; su sobreexpresión altera el equilibrio de señales intracelulares, afectando procesos de diferenciación y crecimiento neuronal.

- **DYRK1A:**  
  Una cinasa que influye en la fosforilación de proteínas clave en la neurogénesis. Su actividad se puede representar mediante modelos cinéticos:

  $$
  v = k_{\text{cat}} \cdot [\text{DYRK1A}] \cdot \frac{[S]}{K_m + [S]},
  $$

  donde $([S]$) es el sustrato fosforilable.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import odeint

# Configuración de estilo
sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (14, 10)

##############################################
# 1. APP: Acumulación de Aβ
##############################################

def app_model(y, t, k_prod, k_deg, factor):
    # Modelo: d[Aβ]/dt = k_prod * (1 + factor) - k_deg * [Aβ]
    return k_prod * (1 + factor) - k_deg * y

t = np.linspace(0, 100, 500)
k_prod = 0.5
k_deg = 0.1

# Condición normal (factor = 0)
sol_normal_APP = odeint(app_model, 0, t, args=(k_prod, k_deg, 0)).flatten()
# Condición SD (factor = 0.5)
sol_sd_APP = odeint(app_model, 0, t, args=(k_prod, k_deg, 0.5)).flatten()

##############################################
# 2. SOD1: Producción de H₂O₂
##############################################

def sod1_reaction(O2_minus, k):
    # Producción de H₂O₂ proporcional a k * [O₂⁻]²
    return k * O2_minus**2

O2_minus = np.linspace(0, 5, 200)
k_normal = 0.1
k_sd = 0.15  # Aumentado en SD

H2O2_normal = sod1_reaction(O2_minus, k_normal)
H2O2_sd = sod1_reaction(O2_minus, k_sd)

##############################################
# 3. S100B: Expresión relativa
##############################################

expr_normal_S100B = 1.0
expr_sd_S100B = 1.5

##############################################
# 4. DYRK1A: Actividad catalítica
##############################################

def dyrk1a_activity(S, k_cat, Km):
    # Modelo: v = (k_cat * S) / (Km + S)
    return (k_cat * S) / (Km + S)

S = np.linspace(0, 20, 200)
k_cat_normal = 2.0
Km = 5.0
k_cat_sd = k_cat_normal * 1.5

v_normal_DYRK1A = dyrk1a_activity(S, k_cat_normal, Km)
v_sd_DYRK1A = dyrk1a_activity(S, k_cat_sd, Km)

##############################################
# Graficación comparativa en subplots
##############################################

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.3)

# (1) APP
axs[0, 0].plot(t, sol_normal_APP, 'b-', label=r'Normal ($\mathrm{factor}=0$)')
axs[0, 0].plot(t, sol_sd_APP, 'r-', label=r'SD ($\mathrm{factor}=0.5$)')
axs[0, 0].fill_between(t, sol_normal_APP, sol_sd_APP, color='gray', alpha=0.3,
                       label='Incremento en producción')
axs[0, 0].set_xlabel('Tiempo')
axs[0, 0].set_ylabel(r'[$\mathrm{A}\beta$]')
axs[0, 0].set_title(r'Acumulación de $\mathrm{A}\beta$ (APP)')
axs[0, 0].legend()

# (2) SOD1
axs[0, 1].plot(O2_minus, H2O2_normal, 'b-', label=r'Normal ($k=0.1$)')
axs[0, 1].plot(O2_minus, H2O2_sd, 'r-', label=r'SD ($k=0.15$)')
axs[0, 1].fill_between(O2_minus, H2O2_normal, H2O2_sd, color='gray', alpha=0.3,
                       label='Incremento en reacción')
axs[0, 1].set_xlabel(r'Concentración $\mathrm{O_{2}^{-}}$')
axs[0, 1].set_ylabel(r'[$\mathrm{H_{2}O_{2}}$]')
axs[0, 1].set_title(r'Producción de $\mathrm{H_{2}O_{2}}$ (SOD1)')
axs[0, 1].legend()

# (3) S100B: gráfico de barras comparativo
labels = ['Normal', 'SD']
expr_levels = [expr_normal_S100B, expr_sd_S100B]
axs[1, 0].bar(labels, expr_levels, color=['skyblue', 'salmon'], edgecolor='black')
# Se simula un intervalo de confianza con fill_between en un área de barras
# Para ello, definimos los bordes inferiores y superiores (ej. ±10%)
lower = [expr * 0.9 for expr in expr_levels]
upper = [expr * 1.1 for expr in expr_levels]
x_pos = np.arange(len(labels))
# Dibujamos líneas de error
axs[1, 0].errorbar(x_pos, expr_levels, yerr=[np.subtract(expr_levels, lower), np.subtract(upper, expr_levels)],
                   fmt='none', ecolor='gray', capsize=5, label='Intervalo ±10%')
axs[1, 0].set_ylabel('Nivel de expresión')
axs[1, 0].set_title('Expresión Relativa de S100B')
axs[1, 0].set_xticks(x_pos)
axs[1, 0].set_xticklabels(labels)
axs[1, 0].legend()

# (4) DYRK1A: Curvas de actividad catalítica
axs[1, 1].plot(S, v_normal_DYRK1A, 'b-', label=r'Normal ($k_{\mathrm{cat}}=2.0$)')
axs[1, 1].plot(S, v_sd_DYRK1A, 'r-', label=r'SD ($k_{\mathrm{cat}}=3.0$)')
axs[1, 1].fill_between(S, v_normal_DYRK1A, v_sd_DYRK1A, color='gray', alpha=0.3,
                       label='Incremento en actividad')
axs[1, 1].set_xlabel('Concentración del sustrato [S]')
axs[1, 1].set_ylabel('Actividad v')
axs[1, 1].set_title('Actividad de DYRK1A')
axs[1, 1].legend()

plt.suptitle(r'Comparativa de Genes Clave y sus Implicaciones', fontsize=16)
plt.show()
```

## III. Sustratos Neurobiológicos del Deterioro Cognitivo en el Síndrome de Down

### III.A. Análisis Cuantitativo de Alteraciones en la Neurogénesis, Sinaptogénesis y Mielinización

**Neurogénesis:**  
Se observa una disminución en la tasa de generación de nuevas neuronas. Si $( N(t) $) representa el número de neuronas en el tiempo $( t $), el proceso de neurogénesis podría modelarse mediante una ecuación diferencial:

$$
\frac{dN}{dt} = \alpha N(t) - \beta N(t),
$$

donde $(\alpha$) es la tasa de proliferación y $(\beta$) es la tasa de diferenciación o pérdida. En SD, la reducción en $(\alpha$) y/o un aumento en $(\beta$) genera un déficit neuronal.

**Sinaptogénesis:**  
La formación de sinapsis, esencial para la transmisión de señales, se puede cuantificar mediante el número de conexiones $(\Gamma$) por neurona. Una reducción en la densidad sináptica $(\rho_s$) afecta la conectividad neuronal:

$$
\rho_s = \frac{\Gamma}{N},
$$

con un valor decreciente en SD, lo que implica menos conexiones para la transferencia de información.

**Mielinización:**  
La velocidad de conducción de los impulsos nerviosos, $( v $), depende del grosor de la vaina de mielina. Este proceso, descrito en términos de física de semiconductores, puede aproximarse mediante la relación:

$$
v \propto \sqrt{d},
$$

donde $( d $) es el espesor de la mielina. En SD, un menor $( d $) se traduce en una reducción en $( v $), afectando la sincronización de la actividad neuronal.

### III.B. Evidencia de Neuroimagen

Los estudios de neuroimagen han mostrado que:

- El volumen cerebral, $( V_{\text{cerebral}} $), es menor en regiones específicas, lo cual se puede cuantificar como:

  $$
  V_{\text{cerebral}} = \int_{\Omega} \rho(\vec{r})\, d\vec{r},
  $$

  donde $(\rho(\vec{r})$) es la densidad neuronal en la región $(\Omega$).

- Las diferencias en el grosor cortical y la integridad de la sustancia blanca confirman la alteración en la conectividad y el procesamiento neuronal.



```python

# Configuración de estilo
sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (10, 6)

##############################################
# III.A. Análisis Cuantitativo de Alteraciones
##############################################

# 1. Neurogénesis: Modelo de crecimiento neuronal
# Ecuación: dN/dt = (alpha - beta)*N(t)
# Se asume que en condición Normal: alpha = 0.05, beta = 0.03 (crecimiento moderado)
# En SD, se reduce la proliferación o se aumenta la pérdida: por ejemplo, alpha = 0.03, beta = 0.04
def neurogenesis_normal(t, N0, alpha=0.05, beta=0.03):
    return N0 * np.exp((alpha - beta) * t)

def neurogenesis_sd(t, N0, alpha=0.03, beta=0.04):
    return N0 * np.exp((alpha - beta) * t)

t = np.linspace(0, 200, 500)
N0 = 100
N_normal = neurogenesis_normal(t, N0)
N_sd = neurogenesis_sd(t, N0)

# Gráfica comparativa para Neurogénesis
fig, ax = plt.subplots()
ax.plot(t, N_normal, 'b-', label='Normal')
ax.plot(t, N_sd, 'r-', label='SD')
ax.fill_between(t, N_normal, N_sd, color='gray', alpha=0.3, label='Diferencia')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Número de Neuronas, N(t)')
ax.set_title('Comparativa de Neurogénesis')
ax.legend()
plt.show()

# 2. Sinaptogénesis: Relación entre número de conexiones (Γ) y neuronas (N)
# Se define la densidad sináptica: ρs = Γ / N
# Ejemplo: Normal: N = 100, Γ = 300 (ρs = 3); SD: N = 90, Γ = 250 (ρs ≈ 2.78)
N_normal_value = 100
Gamma_normal = 300
rho_normal = Gamma_normal / N_normal_value

N_sd_value = 90
Gamma_sd = 250
rho_sd = Gamma_sd / N_sd_value

fig, ax = plt.subplots()
labels = ['Normal', 'SD']
rho_values = [rho_normal, rho_sd]
bars = ax.bar(labels, rho_values, color=['skyblue', 'salmon'], edgecolor='black')
ax.set_ylabel(r'Densidad Sináptica, $\rho_s$')
ax.set_title('Comparativa de Sinaptogénesis')
# Agregamos texto explicativo en cada barra
for bar, val in zip(bars, rho_values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height*1.02, f'{val:.2f}', ha='center', va='bottom')
plt.show()

# 3. Mielinización: Relación entre espesor de mielina (d) y velocidad de conducción (v ∝ √d)
# Ejemplo: Normal: d = 4 (v = 2), SD: d = 3 (v ≈ 1.73)
d_normal = 4
v_normal = np.sqrt(d_normal)

d_sd = 3
v_sd = np.sqrt(d_sd)

fig, ax = plt.subplots()
conditions = ['Normal', 'SD']
v_values = [v_normal, v_sd]
bars = ax.bar(conditions, v_values, color=['skyblue', 'salmon'], edgecolor='black')
ax.set_ylabel('Velocidad de Conducción, v')
ax.set_title('Comparativa de Mielinización')
for bar, val in zip(bars, v_values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height*1.02, f'{val:.2f}', ha='center', va='bottom')
plt.show()

##############################################
# III.B. Evidencia de Neuroimagen
##############################################

# Se aproxima el volumen cerebral como: V = (4/3)*π*r³ * ρ, donde ρ es la densidad neuronal promedio.
# Ejemplo: Normal: r = 10, ρ = 1.0; SD: r = 9, ρ = 0.9
r_normal = 10
rho_normal_brain = 1.0
V_normal = (4/3)*np.pi*(r_normal**3) * rho_normal_brain

r_sd = 9
rho_sd_brain = 0.9
V_sd = (4/3)*np.pi*(r_sd**3) * rho_sd_brain

fig, ax = plt.subplots()
conditions = ['Normal', 'SD']
volumes = [V_normal, V_sd]
bars = ax.bar(conditions, volumes, color=['skyblue', 'salmon'], edgecolor='black')
ax.set_ylabel(r'Volumen Cerebral, $V_{\mathrm{cerebral}}$ (u.a.)')
ax.set_title('Comparativa de Volumen Cerebral (Neuroimagen)')
for bar, vol in zip(bars, volumes):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height*1.02, f'{vol:.0f}', ha='center', va='bottom')
plt.show()
```

## IV. Sistemas Neuroquímicos y Función Cognitiva en el Síndrome de Down

### IV.A. Modelado Matemático de la Desregulación de Neurotransmisores

En el SD, se observan alteraciones en los sistemas colinérgico, dopaminérgico y serotoninérgico. Estos sistemas pueden modelarse mediante ecuaciones diferenciales que describen la dinámica de los neurotransmisores. Por ejemplo, para el sistema colinérgico:

$$
\frac{d[ACh]}{dt} = R_{\text{prod}} - R_{\text{deg}}[ACh],
$$

donde $([ACh]$) es la concentración de acetilcolina, $(R_{\text{prod}}$) es la tasa de producción y $(R_{\text{deg}}$) es la constante de degradación. Alteraciones en estos parámetros se reflejan en déficits en memoria y atención.

En el sistema dopaminérgico, la función $( D(t) $) de la dopamina puede verse afectada según:

$$
\frac{dD}{dt} = k_{\text{syn}} - k_{\text{uptake}}D,
$$

donde $( k_{\text{syn}} $) es la tasa de síntesis y $( k_{\text{uptake}} $) la tasa de recaptación. La reducción en $( D(t) $) se asocia a dificultades en funciones ejecutivas.

### IV.B. Impacto de los Desequilibrios Neuroquímicos

La interacción entre estos sistemas puede describirse mediante modelos acoplados. Por ejemplo, la interacción entre los sistemas colinérgico y dopaminérgico se puede modelar con un sistema de ecuaciones:

$$
\begin{aligned}
\frac{d[ACh]}{dt} &= R_{\text{ACh}} - k_{\text{ACh}}[ACh] - \gamma D[ACh], \\
\frac{dD}{dt} &= R_{D} - k_{D}D - \delta [ACh]D,
\end{aligned}
$$

donde $(\gamma$) y $(\delta$) son constantes que cuantifican la interacción entre ambos sistemas. Dichas interacciones explican cómo una alteración en un sistema puede desencadenar un efecto en cascada en el otro, afectando la función cognitiva.


```python

############################################################
# IV.A. Modelado Matemático de la Desregulación de Neurotransmisores
############################################################

# Sistema Colinérgico: d[ACh]/dt = R_prod - R_deg * [ACh]
def cholinergic_model(ACh, t, R_prod, R_deg):
    return R_prod - R_deg * ACh

# Sistema Dopaminérgico: dD/dt = k_syn - k_uptake * D
def dopaminergic_model(D, t, k_syn, k_uptake):
    return k_syn - k_uptake * D

# Tiempo de simulación
t = np.linspace(0, 100, 500)

# Parámetros para condición Normal
R_prod_normal = 1.0
R_deg_normal  = 0.1
k_syn_normal  = 1.0
k_uptake_normal = 0.1

# Parámetros para condición SD (hipotéticos: menor síntesis o mayor degradación)
R_prod_sd = 0.8       # Menor producción de ACh
R_deg_sd  = 0.12      # Mayor degradación de ACh
k_syn_sd  = 0.8       # Menor síntesis de dopamina
k_uptake_sd = 0.12    # Mayor recaptación de dopamina

# Simulación del sistema colinérgico
ACh_normal = odeint(cholinergic_model, 0, t, args=(R_prod_normal, R_deg_normal)).flatten()
ACh_sd     = odeint(cholinergic_model, 0, t, args=(R_prod_sd, R_deg_sd)).flatten()

# Simulación del sistema dopaminérgico
D_normal = odeint(dopaminergic_model, 0, t, args=(k_syn_normal, k_uptake_normal)).flatten()
D_sd     = odeint(dopaminergic_model, 0, t, args=(k_syn_sd, k_uptake_sd)).flatten()

# Graficación comparativa para sistemas individuales
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Gráfica del sistema colinérgico
axs[0].plot(t, ACh_normal, 'b-', label='Normal: $R_{prod}=1.0$, $R_{deg}=0.1$')
axs[0].plot(t, ACh_sd, 'r-', label='SD: $R_{prod}=0.8$, $R_{deg}=0.12$')
axs[0].fill_between(t, ACh_normal, ACh_sd, color='gray', alpha=0.3, label='Diferencia')
axs[0].set_xlabel('Tiempo')
axs[0].set_ylabel(r'Concentración $[ACh]$')
axs[0].set_title('Sistema Colinérgico')
axs[0].legend()

# Gráfica del sistema dopaminérgico
axs[1].plot(t, D_normal, 'b-', label='Normal: $k_{syn}=1.0$, $k_{uptake}=0.1$')
axs[1].plot(t, D_sd, 'r-', label='SD: $k_{syn}=0.8$, $k_{uptake}=0.12$')
axs[1].fill_between(t, D_normal, D_sd, color='gray', alpha=0.3, label='Diferencia')
axs[1].set_xlabel('Tiempo')
axs[1].set_ylabel(r'Concentración $[D]$')
axs[1].set_title('Sistema Dopaminérgico')
axs[1].legend()

plt.suptitle(r'Comparativa de Sistemas Neuroquímicos: Normal vs SD', fontsize=16)
plt.show()

############################################################
# IV.B. Impacto de los Desequilibrios Neuroquímicos: Modelo Acoplado
############################################################

def coupled_model(y, t, R_ACh, k_ACh, R_D, k_D, gamma, delta):
    """
    Modelo acoplado para describir la interacción entre los sistemas colinérgico y dopaminérgico:
    
      d[ACh]/dt = R_ACh - k_ACh * [ACh] - γ * [D] * [ACh]
      d[D]/dt   = R_D - k_D * [D] - δ * [ACh] * [D]
    
    """
    ACh, D = y
    dACh_dt = R_ACh - k_ACh * ACh - gamma * D * ACh
    dD_dt   = R_D - k_D * D - delta * ACh * D
    return [dACh_dt, dD_dt]

# Parámetros para condición Normal
params_normal = {
    'R_ACh': 1.0,
    'k_ACh': 0.1,
    'R_D': 1.0,
    'k_D': 0.1,
    'gamma': 0.05,
    'delta': 0.05
}

# Parámetros para condición SD (hipotéticos: mayores interacciones que agravan el déficit)
params_sd = {
    'R_ACh': 0.8,     # Menor producción de ACh
    'k_ACh': 0.12,    # Mayor degradación
    'R_D': 0.8,       # Menor síntesis de dopamina
    'k_D': 0.12,      # Mayor recaptación
    'gamma': 0.08,    # Interacción ACh-D disminuida o aumentada (hipotético)
    'delta': 0.08
}

# Tiempo y condiciones iniciales
t_coupled = np.linspace(0, 50, 500)
y0 = [0, 0]  # Suponemos que inicialmente ambas concentraciones son 0

# Simulación en condiciones normales
sol_normal = odeint(coupled_model, y0, t_coupled, args=(params_normal['R_ACh'], params_normal['k_ACh'],
                                                          params_normal['R_D'], params_normal['k_D'],
                                                          params_normal['gamma'], params_normal['delta']))
# Simulación en condición SD
sol_sd = odeint(coupled_model, y0, t_coupled, args=(params_sd['R_ACh'], params_sd['k_ACh'],
                                                      params_sd['R_D'], params_sd['k_D'],
                                                      params_sd['gamma'], params_sd['delta']))

fig, ax = plt.subplots()
# Graficamos [ACh] en ambas condiciones
ax.plot(t_coupled, sol_normal[:, 0], 'b--', label='Normal [ACh]')
ax.plot(t_coupled, sol_sd[:, 0], 'r--', label='SD [ACh]')
# Graficamos [D] en ambas condiciones
ax.plot(t_coupled, sol_normal[:, 1], 'b-', label='Normal [D]')
ax.plot(t_coupled, sol_sd[:, 1], 'r-', label='SD [D]')
# Rellenamos el área entre condiciones para [ACh] (opcional)
ax.fill_between(t_coupled, sol_normal[:, 0], sol_sd[:, 0], color='gray', alpha=0.3,
                label='Diferencia [ACh]')
ax.fill_between(t_coupled, sol_normal[:, 1], sol_sd[:, 1], color='gray', alpha=0.3,
                label='Diferencia [D]')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Concentración')
ax.set_title('Comparativa del Modelo Acoplado: Sistemas Neuroquímicos')
ax.legend(loc='upper right')
plt.show()
```

## V. Base Neurocientífica de las Pruebas Cognitivas en el Conjunto de Datos

Cada prueba cognitiva se fundamenta en la actividad de regiones cerebrales específicas y en la dinámica de señales bioeléctricas. Por ejemplo:

- **PPVT (Prueba de Vocabulario en Imágenes Peabody):**  
  Se asocia a la activación en regiones del lóbulo temporal, donde la respuesta neuronal se puede modelar mediante funciones de activación $( f(x) $) de tipo sigmoide:

  $$
  f(x) = \frac{1}{1 + e^{-x}},
  $$

  que permiten cuantificar la respuesta frente a estímulos lingüísticos.

- **Matrices Progresivas de Raven:**  
  Evalúan la capacidad de razonamiento abstracto, la cual puede modelarse con redes neuronales donde cada neurona sigue la dinámica:

  $$
  a_j = f\left(\sum_i w_{ij} x_i + b_j\right),
  $$

  representando la integración de señales de diferentes regiones corticales.

- **Pruebas de Memoria de Trabajo (WM):**  
  Estas pruebas exploran la capacidad de mantener y manipular información. La dinámica de la memoria de trabajo puede describirse con modelos de osciladores acoplados, donde la estabilidad de las oscilaciones determina la persistencia de la información:

  $$
  \dot{\theta}_i = \omega_i + \sum_j K_{ij} \sin(\theta_j - \theta_i),
  $$

  en donde $(\theta_i$) representa la fase de la actividad neuronal y $(K_{ij}$) la fuerza de acoplamiento entre neuronas.



```python

##############################################
# 1. PPVT: Función de Activación Sigmoide
##############################################
x = np.linspace(-10, 10, 400)
f_x = 1 / (1 + np.exp(-x))

# Graficamos la función sigmoide
fig, ax = plt.subplots()
ax.plot(x, f_x, 'b-', label=r'$f(x)=\frac{1}{1+e^{-x}}$')
ax.fill_between(x, f_x, color='blue', alpha=0.2)
ax.set_xlabel('Estímulo')
ax.set_ylabel('Activación')
ax.set_title('PPVT: Función de Activación Sigmoide\n(Estimulación en el lóbulo temporal)')
ax.legend()
plt.show()

##############################################
# 2. Matrices Progresivas de Raven: Modelo de Activación Neuronal
##############################################
def neuron_activation(x, w, b):
    return 1 / (1 + np.exp(-(w * x + b)))

x_raven = np.linspace(-10, 10, 400)
activation1 = neuron_activation(x_raven, w=1.0, b=0.0)
activation2 = neuron_activation(x_raven, w=1.5, b=-1.0)

fig, ax = plt.subplots()
ax.plot(x_raven, activation1, 'g-', label=r'Neuron 1: $w=1.0,\;b=0.0$')
ax.plot(x_raven, activation2, 'm-', label=r'Neuron 2: $w=1.5,\;b=-1.0$')
ax.fill_between(x_raven, activation1, activation2, color='gray', alpha=0.3, 
                label='Diferencia de activación')
ax.set_xlabel('Estimulación')
ax.set_ylabel('Respuesta neuronal')
ax.set_title('Matrices Progresivas de Raven: Modelo de Activación Neuronal')
ax.legend()
plt.show()

##############################################
# 3. Pruebas de Memoria de Trabajo (WM): Modelo Kuramoto
##############################################
def kuramoto(theta, t, omega, K, N):
    dtheta = np.zeros(N)
    for i in range(N):
        dtheta[i] = omega[i] + (K / N) * np.sum(np.sin(theta - theta[i]))
    return dtheta

# Simulación para 10 osciladores
N = 10
t_kuramoto = np.linspace(0, 20, 300)
np.random.seed(42)
omega = np.random.uniform(0.8, 1.2, N)
theta0 = np.random.uniform(0, 2 * np.pi, N)
theta_t = odeint(kuramoto, theta0, t_kuramoto, args=(omega, 1.0, N))

fig, ax = plt.subplots()
for i in range(N):
    ax.plot(t_kuramoto, theta_t[:, i], label=f'Osc {i+1}')
ax.set_xlabel('Tiempo')
ax.set_ylabel(r'Fase $\theta$')
ax.set_title('Pruebas de Memoria de Trabajo: Dinámica de Osciladores (Modelo Kuramoto)')
ax.legend(ncol=2, bbox_to_anchor=(1.05, 1.0))
plt.tight_layout()
plt.show()
```

## VI. Ingeniería Biomédica en la Evaluación y Diagnóstico de Deficiencias Cognitivas

### VI.A. Aplicación de Biosensores y Química Analítica

Los biosensores utilizan principios de química analítica para detectar biomarcadores. Por ejemplo, se emplean superficies funcionalizadas con nanopartículas de oro (AuNPs) que interactúan con proteínas mal plegadas mediante enlaces covalentes o de coordinación, representados por:

$$
\mathrm{AuNP{-}R} + \mathrm{Protein} \longrightarrow \mathrm{AuNP{-}Protein},
$$

donde $( \mathrm{R} $) denota el ligando funcional en la superficie del biosensor. Estas interacciones permiten medir la concentración de proteínas patológicas con alta sensibilidad y especificidad.

### VI.B. Procesamiento de Señales Biológicas y Modelos Matemáticos

El análisis de señales bioeléctricas, como el EEG, se basa en la transformada de Fourier, definida por:

$$
X(f) = \int_{-\infty}^{\infty} x(t)e^{-i2\pi ft} \, dt,
$$

lo que permite identificar patrones espaciales y temporales que correlacionan con estados cognitivos. La aplicación de algoritmos de machine learning a estas transformadas facilita la detección temprana del deterioro cognitivo.

### VI.C. Integración de Datos Multidimensionales

La fusión de datos de neuroimagen, señales bioeléctricas y biomarcadores químicos se aborda mediante técnicas de análisis multivariante, donde se puede utilizar el álgebra lineal para encontrar la representación de menor dimensión que preserve la información crítica:

$$
\mathbf{Y} = \mathbf{W}^T \mathbf{X},
$$

donde $(\mathbf{X}$) es la matriz de datos originales, $(\mathbf{W}$) la matriz de proyecciones, y $(\mathbf{Y}$) la representación reducida que alimenta los modelos predictivos.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import odeint
from scipy.fft import fft, fftfreq
from sklearn.decomposition import PCA

# Configuración de estilo
sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (10, 6)

##############################################
# VI.A. Aplicación de Biosensores y Química Analítica
##############################################
def sensor_response(P, R_max, Km):
    """
    Modelo de respuesta del sensor (tipo Michaelis–Menten):
      R(P) = R_max * P / (Km + P)
    """
    return R_max * P / (Km + P)

# Se simulan dos condiciones:
# - Normal: con Km = 20
# - SD: se asume mayor sensibilidad (Km = 15)
P_values = np.linspace(0, 100, 300)
R_max = 100
Km_normal = 20
Km_sd = 15

response_normal = sensor_response(P_values, R_max, Km_normal)
response_sd = sensor_response(P_values, R_max, Km_sd)

fig, ax = plt.subplots()
ax.plot(P_values, response_normal, 'b-', label=r'Normal: $K_m=20$')
ax.plot(P_values, response_sd, 'r-', label=r'SD: $K_m=15$')
ax.fill_between(P_values, response_normal, response_sd, color='gray', alpha=0.3,
                label='Incremento en sensibilidad')
ax.set_xlabel('Concentración de proteína [P]')
ax.set_ylabel('Respuesta del sensor')
ax.set_title(r'Respuesta del Biosensor: $\mathrm{AuNP-R} + \mathrm{Protein} \rightarrow \mathrm{AuNP-Protein}$')
ax.legend()
plt.show()

##############################################
# VI.B. Procesamiento de Señales Biológicas y Modelos Matemáticos
##############################################
# Simulación de dos señales EEG sintéticas
t_signal = np.linspace(0, 2, 400)
# Señal Normal: pico en 10 Hz
signal_normal = np.sin(2*np.pi*10*t_signal) + 0.5*np.sin(2*np.pi*20*t_signal)
# Señal SD: pico en 6 Hz (hipotético)
signal_sd = np.sin(2*np.pi*6*t_signal) + 0.5*np.sin(2*np.pi*12*t_signal)

# Cálculo de la transformada de Fourier
yf_normal = fft(signal_normal)
yf_sd = fft(signal_sd)
xf = fftfreq(len(t_signal), t_signal[1]-t_signal[0])
xf_pos = xf[:len(xf)//2]
mag_normal = np.abs(yf_normal)[:len(xf)//2]
mag_sd = np.abs(yf_sd)[:len(xf)//2]

fig, ax = plt.subplots()
ax.plot(xf_pos, mag_normal, 'b-', label='Normal')
ax.plot(xf_pos, mag_sd, 'r-', label='SD')
ax.fill_between(xf_pos, mag_normal, mag_sd, color='gray', alpha=0.3,
                label='Diferencia espectral')
ax.set_xlabel('Frecuencia (Hz)')
ax.set_ylabel('Magnitud')
ax.set_title('Transformada de Fourier: Señales EEG Normal vs SD')
ax.legend()
plt.show()

##############################################
# VI.C. Integración de Datos Multidimensionales
##############################################
def generate_data(n_samples, n_features, shift=0.0):
    """
    Genera datos aleatorios (distribución normal).
    El parámetro shift simula diferencias (por ejemplo, en SD).
    """
    np.random.seed(0)
    return np.random.randn(n_samples, n_features) + shift

# Datos Normal y SD
X_normal = generate_data(200, 5, shift=0.0)
X_sd = generate_data(200, 5, shift=0.5)  # Desplazamiento para SD

# Aplicamos PCA para reducir a 2 dimensiones
pca_normal = PCA(n_components=2)
X_normal_reduced = pca_normal.fit_transform(X_normal)

pca_sd = PCA(n_components=2)
X_sd_reduced = pca_sd.fit_transform(X_sd)

fig, ax = plt.subplots()
ax.scatter(X_normal_reduced[:, 0], X_normal_reduced[:, 1], c='blue', alpha=0.6, label='Normal')
ax.scatter(X_sd_reduced[:, 0], X_sd_reduced[:, 1], c='red', alpha=0.6, label='SD')
ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')
ax.set_title('Proyección PCA: Integración de Datos Multidimensionales')
ax.legend()
plt.show()

```

## VII. Modelos Matemáticos y Computacionales en la Neurociencia del Desarrollo Cognitivo

Se emplean simulaciones computacionales para modelar procesos complejos del desarrollo cognitivo. Por ejemplo, el desarrollo léxico puede simularse mediante modelos de redes de asociación, donde la probabilidad de formar una conexión entre dos conceptos $( A $) y $( B $) se define como:

$$
P(A \leftrightarrow B) = \frac{e^{\kappa \, \text{sim}(A,B)}}{\sum_{C} e^{\kappa \, \text{sim}(A,C)}},
$$

donde $(\text{sim}(A,B)$) es una función de similitud y $(\kappa$) un parámetro de dispersión.

Asimismo, modelos de redes neuronales in silico permiten simular la topología neuronal y la distribución de sinapsis, proporcionando una herramienta para comprender cómo las alteraciones estructurales afectan la función cerebral en SD.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import odeint
import networkx as nx

# Configuración de estilo
sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (10, 6)

##############################################
# Parte 1: Modelo de Asociación Léxica
##############################################

def lexical_association(sim, kappa, n=5, sim0=0.5):
    """
    Modelo simplificado:
    P(A <-> B) = exp(kappa * sim) / ( exp(kappa * sim) + (n-1)*exp(kappa * sim0) )
    
    Parámetros:
      - sim: similitud entre A y B (valor entre 0 y 1)
      - kappa: parámetro de dispersión
      - n: número total de conceptos (se asume que para los demás la similitud es sim0)
      - sim0: similitud promedio con los otros conceptos
    """
    num = np.exp(kappa * sim)
    den = num + (n - 1) * np.exp(kappa * sim0)
    return num / den

sim_values = np.linspace(0, 1, 200)
kappa_values = [0.5, 1.0, 2.0, 5.0]

fig, ax = plt.subplots()
for kappa in kappa_values:
    P = lexical_association(sim_values, kappa)
    ax.plot(sim_values, P, label=f'κ = {kappa}')
ax.set_xlabel('Similitud $\mathrm{sim}(A,B)$')
ax.set_ylabel('Probabilidad de conexión $P(A\\leftrightarrow B)$')
ax.set_title('Modelo de Asociación Léxica')
ax.legend()
plt.show()

##############################################
# Parte 2: Simulación de una Red Neuronal In Silico
##############################################
# Para ilustrar la topología y distribución de sinapsis,
# generamos una red aleatoria donde cada nodo representa un concepto y
# la similitud se asigna aleatoriamente para ponderar las conexiones.

# Número de nodos (conceptos)
num_nodes = 15
# Probabilidad de conexión entre nodos
p_connection = 0.3

# Creamos la red usando el modelo de Erdős-Rényi
G = nx.erdos_renyi_graph(num_nodes, p_connection, seed=42)

# Asignamos a cada arista un "peso" de similitud (valor entre 0 y 1)
for u, v in G.edges():
    G.edges[u, v]['weight'] = np.random.uniform(0.3, 1.0)

# Visualización: El grosor de la arista es proporcional al peso
pos = nx.spring_layout(G, seed=42)
weights = [G.edges[u, v]['weight'] for u, v in G.edges()]
edge_widths = [w * 3 for w in weights]  # Ajustamos la escala para visualizar

plt.figure()
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(G, pos, width=edge_widths, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
plt.title('Simulación de Red Neuronal In Silico\n(Distribución de Sinapsis y Topología)')
plt.axis('off')
plt.show()
```

## VIII. Avances en Biotecnología y Terapia Génica para Mitigar las Deficiencias Cognitivas

Aunque el foco del proyecto es la detección y clasificación, se exploran estrategias terapéuticas basadas en biotecnología y terapia génica. Algunas aproximaciones incluyen:

- **Molécula AEF0217:**  
  Su eficacia se evalúa mediante estudios farmacocinéticos y farmacodinámicos, donde la concentración plasmática $( C(t) $) sigue un modelo de dos compartimentos:

  $$
  \frac{dC}{dt} = -k_{el} \, C(t),
  $$

  con $(k_{el}$) siendo la constante de eliminación.

- **Galato de Epigalocatequina (EGCG):**  
  Este polifenol ($(\mathrm{C}_{22}\mathrm{H}_{18}\mathrm{O}_{11}$)) modula la actividad de DYRK1A, y su efecto puede cuantificarse mediante curvas de dosis-respuesta descritas por:

  $$
  E = \frac{E_{\max} \, [\text{EGCG}]}{EC_{50} + [\text{EGCG}]},
  $$

  donde $(E$) es la eficacia, $(E_{\max}$) la eficacia máxima y $(EC_{50}$) la concentración a la mitad del efecto máximo.

- **Terapia Génica In Utero:**  
  La normalización de la copia de genes críticos, como APP y DYRK1A, se modela considerando la transferencia de vectores virales y la integración génica, utilizando modelos probabilísticos para predecir la eficacia y seguridad del procedimiento.

- **Fármaco Experimental ISRIB:**  
  Este fármaco actúa modulando la actividad de factores de estrés celular. Su impacto en la memoria se evalúa mediante modelos de plasticidad sináptica, que pueden incluir ecuaciones diferenciales para la potenciación a largo plazo (LTP):

  $$
  \frac{d\text{LTP}}{dt} = \alpha \, \text{Input} - \beta \, \text{LTP},
  $$

  donde $(\alpha$) y $(\beta$) representan tasas de inducción y decaimiento, respectivamente.



```python

##############################################
# 1. Molécula AEF0217: Concentración Plasmática
##############################################
def concentration_model(C, t, k_el):
    # dC/dt = - k_el * C  =>  C(t) = C0 * exp(- k_el * t)
    return -k_el * C

t = np.linspace(0, 100, 500)
C0 = 100

k_el_normal = 0.1
k_el_sd = 0.15

C_normal = odeint(concentration_model, C0, t, args=(k_el_normal,)).flatten()
C_sd = odeint(concentration_model, C0, t, args=(k_el_sd,)).flatten()

##############################################
# 2. Galato de Epigalocatequina (EGCG): Dosis-Respuesta
##############################################
def dose_response(EGCG, E_max, EC50):
    return E_max * EGCG / (EC50 + EGCG)

EGCG_vals = np.linspace(0, 100, 300)
E_max = 100
EC50_normal = 10
EC50_sd = 7  # En SD se asume mayor eficacia (menor EC50)

E_normal = dose_response(EGCG_vals, E_max, EC50_normal)
E_sd = dose_response(EGCG_vals, E_max, EC50_sd)

##############################################
# 3. Terapia Génica In Utero: Probabilidad de Éxito
##############################################
def success_probability(x, a, x0):
    # Modelo logístico: P = 1/(1 + exp(-a*(x - x0)))
    return 1 / (1 + np.exp(-a * (x - x0)))

x_vals = np.linspace(0, 100, 300)
# En Normal (sin terapia), la integración es menos eficiente
P_normal = success_probability(x_vals, a=0.1, x0=50)
# Con terapia génica, se desplaza el punto de inflexión
P_therapy = success_probability(x_vals, a=0.1, x0=45)

##############################################
# 4. Fármaco Experimental ISRIB: Dinámica de LTP
##############################################
def ltp_model(LTP, t, alpha, beta, Input):
    # Solución con LTP(0)=0: LTP(t) = (alpha*Input/beta)*(1 - exp(-beta*t))
    return alpha * Input - beta * LTP

t_ltp = np.linspace(0, 100, 500)
Input_val = 2.0

alpha_normal = 0.5
beta_normal = 0.05
alpha_sd = 0.3
beta_sd = 0.07

LTP_normal = odeint(ltp_model, 0, t_ltp, args=(alpha_normal, beta_normal, Input_val)).flatten()
LTP_sd = odeint(ltp_model, 0, t_ltp, args=(alpha_sd, beta_sd, Input_val)).flatten()

##############################################
# Graficación comparativa en subplots
##############################################
fig, axs = plt.subplots(2, 2, figsize=(14, 12))
plt.subplots_adjust(hspace=0.3, wspace=0.3)

# (1) Molécula AEF0217: Concentración Plasmática
axs[0, 0].plot(t, C_normal, 'b-', label=r'Normal: $k_{el}=0.1$')
axs[0, 0].plot(t, C_sd, 'r-', label=r'SD: $k_{el}=0.15$')
axs[0, 0].fill_between(t, C_normal, C_sd, color='gray', alpha=0.3, label='Diferencia en eliminación')
axs[0, 0].set_xlabel('Tiempo')
axs[0, 0].set_ylabel(r'$C(t)$')
axs[0, 0].set_title('Modelo Farmacocinético: Molécula AEF0217')
axs[0, 0].legend()

# (2) EGCG: Curva de Dosis-Respuesta
axs[0, 1].plot(EGCG_vals, E_normal, 'b-', label=r'Normal: $EC_{50}=10$')
axs[0, 1].plot(EGCG_vals, E_sd, 'r-', label=r'SD: $EC_{50}=7$')
axs[0, 1].fill_between(EGCG_vals, E_normal, E_sd, color='gray', alpha=0.3, label='Diferencia en eficacia')
axs[0, 1].set_xlabel('Concentración de EGCG')
axs[0, 1].set_ylabel('Eficacia E')
axs[0, 1].set_title('Curva de Dosis-Respuesta para EGCG')
axs[0, 1].legend()

# (3) Terapia Génica In Utero: Probabilidad de Éxito
axs[1, 0].plot(x_vals, P_normal, 'b-', label=r'Normal (sin terapia): $x_0=50$')
axs[1, 0].plot(x_vals, P_therapy, 'r-', label=r'Terapia Génica: $x_0=45$')
axs[1, 0].fill_between(x_vals, P_normal, P_therapy, color='gray', alpha=0.3,
                       label='Incremento de éxito')
axs[1, 0].set_xlabel('Eficiencia de Integración (%)')
axs[1, 0].set_ylabel('Probabilidad de Éxito')
axs[1, 0].set_title('Modelo Probabilístico de Terapia Génica In Utero')
axs[1, 0].legend()

# (4) ISRIB: Dinámica de LTP
axs[1, 1].plot(t_ltp, LTP_normal, 'b-', label=r'Normal: $\alpha=0.5,\;\beta=0.05$')
axs[1, 1].plot(t_ltp, LTP_sd, 'r-', label=r'SD: $\alpha=0.3,\;\beta=0.07$')
axs[1, 1].fill_between(t_ltp, LTP_normal, LTP_sd, color='gray', alpha=0.3,
                       label='Diferencia en plasticidad')
axs[1, 1].set_xlabel('Tiempo')
axs[1, 1].set_ylabel('Nivel de LTP')
axs[1, 1].set_title('Dinámica de Potenciación a Largo Plazo (ISRIB)')
axs[1, 1].legend()

plt.suptitle('Avances en Biotecnología y Terapia Génica para Mitigar las Deficiencias Cognitivas', fontsize=16)
plt.show()
```

## ¿Qué Transformaciones se Van a Usar?

Se utilizarán las siguientes transformaciones para preparar los datos y asegurar que las variables sean adecuadas para el modelado con regresión logística y otros métodos de clasificación:

- **One-Hot Encoding:**  
  Cada variable categórica se convertirá en un vector binario. Si una variable tiene $K$ categorías, cada muestra se representará como un vector $v \in \{0,1\}^K$. Esto es esencial para incorporar variables cualitativas en modelos que requieren entrada numérica.

- **Normalización:**  
  La normalización consiste en escalar los datos numéricos al rango $[0,1]$. La fórmula utilizada es:
  
  $$
  x' = \frac{x - \min(x)}{\max(x) - \min(x)}
  $$
  
  Esta transformación es útil cuando se desea preservar la distribución original de los datos sin alterar su forma, facilitando la comparación entre variables con diferentes escalas.

- **Estandarización:**  
  La estandarización transforma los datos para que tengan una media de $0$ y una desviación estándar de $1$, mediante la fórmula:
  
  $$
  x' = \frac{x - \mu}{\sigma}
  $$
  
  Esta técnica es recomendable cuando se espera que las variables tengan una distribución aproximadamente normal y se desea centrar y escalar los datos de manera que las diferencias de escala y dispersión sean eliminadas.

- **Decisión para Este Proyecto:**  
  En el contexto del presente proyecto, dado que las variables numéricas (como $Age\_months$, puntajes en pruebas cognitivas, etc.) provienen de diferentes escalas y pueden presentar varianzas significativas, se optará por la **estandarización**. Esta transformación ayudará a que los algoritmos de clasificación (regresión logística, SVM, MLP) operen de manera óptima al tratar variables con media $0$ y varianza $1$, facilitando la convergencia durante el entrenamiento y reduciendo el impacto de posibles valores atípicos.

- **Ingeniería de Características:**  
  Se explorarán combinaciones de variables, tales como cocientes o interacciones, para capturar relaciones complejas que puedan mejorar la capacidad predictiva del modelo. Esta etapa puede implicar la creación de nuevas variables derivadas de las existentes.

- **Bootstrap:**  
  Se aplicará la técnica de bootstrap para obtener estimaciones robustas de la distribución de las métricas de evaluación. El bootstrap consiste en muestrear con reemplazo el dataset original y recalcular los estimadores (por ejemplo, accuracy, F1-score) para obtener intervalos de confianza y estimar la variabilidad de los resultados. Formalmente, si se tiene una estadística $T$ calculada sobre una muestra de tamaño $n$, se generan múltiples muestras bootstrap $\{x_1^*, x_2^*, \dots, x_n^*\}$ y se calcula la distribución empírica de $T$ para estimar su sesgo y varianza.

Estas transformaciones y técnicas garantizan que los datos se encuentren en un formato óptimo para el análisis y la predicción del nivel de déficit cognitivo, asegurando la robustez y la fiabilidad del modelo resultante.

```python
import itertools
import numpy as np
import pandas as pd
import warnings
from sklearn.exceptions import ConvergenceWarning

# Suprimir warnings de convergencia para no saturar la salida durante la optimización
warnings.filterwarnings("ignore", category=ConvergenceWarning)

# Importación de módulos de scikit-learn
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, GridSearchCV, cross_val_predict
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             balanced_accuracy_score, roc_auc_score, confusion_matrix)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
df_features = df.drop(columns=['deficit_cognitivo'])
y = df['deficit_cognitivo']
numeric_cols = df_features.columns.tolist()

preprocessor = ColumnTransformer(transformers=[
    ('scale', StandardScaler(), numeric_cols)
])

cv = KFold(n_splits=5, shuffle=True, random_state=50)
```

```python
# Regresión Logística Ridge (L2)
pipe_lr_ridge = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('clf', LogisticRegression(penalty='l2', solver='liblinear', random_state=50))
])
param_grid_lr_ridge = {
    'clf__C': [0.01, 0.1, 1, 10, 100]
}

# Regresión Logística Lasso (L1)
pipe_lr_lasso = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('clf', LogisticRegression(penalty='l1', solver='liblinear', random_state=50))
])
param_grid_lr_lasso = {
    'clf__C': [0.01, 0.1, 1, 10, 100]
}

# SVC con kernel RBF
pipe_svc_rbf = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('clf', SVC(kernel='rbf', probability=True, random_state=50))
])
param_grid_svc_rbf = {
    'clf__C': [0.1, 1, 10],
    'clf__gamma': [0.001, 0.01, 0.1]
}

# SVC con kernel Poly
pipe_svc_poly = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('clf', SVC(kernel='poly', probability=True, random_state=50))
])
param_grid_svc_poly = {
    'clf__C': [0.1, 1, 10],
    'clf__gamma': [0.001, 0.01, 0.1],
    'clf__degree': [2, 3, 4]
}

# SVC con kernel Sigmoid
pipe_svc_sigmoid = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('clf', SVC(kernel='sigmoid', probability=True, random_state=50))
])
param_grid_svc_sigmoid = {
    'clf__C': [0.1, 1, 10],
    'clf__gamma': [0.001, 0.01, 0.1]
}

# MLP
pipe_mlp = Pipeline([
    ('preprocessor', preprocessor),
    ('pca', PCA()),
    ('clf', MLPClassifier(random_state=50, max_iter=1000, tol=1e-4))
])
param_grid_mlp = {
    'clf__hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'clf__alpha': [1e-5, 1e-3, 1e-1],
    'clf__learning_rate_init': [1e-4, 1e-3, 1e-2]
}
```

```python
def run_grid_search(pipe, param_grid, X, y, cv):
    gs = GridSearchCV(
        estimator=pipe,
        param_grid=param_grid,
        scoring='accuracy',
        cv=cv,
        n_jobs=1  # n_jobs=1 para conservar recursos
    )
    gs.fit(X, y)
    return gs.best_estimator_, gs.best_params_, gs.best_score_

def evaluate_model(name, model, X, y, cv):
    y_pred = cross_val_predict(model, X, y, cv=cv, n_jobs=1)
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y, y_pred, average='weighted', zero_division=0)
    bal_acc = balanced_accuracy_score(y, y_pred)
    try:
        y_proba = cross_val_predict(model, X, y, cv=cv, n_jobs=1, method='predict_proba')
        classes = list(np.unique(y))
        y_bin = label_binarize(y, classes=classes)
        roc_auc = roc_auc_score(y_bin, y_proba, multi_class='ovr')
    except:
        roc_auc = np.nan
    cm = confusion_matrix(y, y_pred)
    return {
        'Model': name,
        'Accuracy': acc,
        'Precision_weighted': prec,
        'Recall_weighted': rec,
        'F1_weighted': f1,
        'Balanced_Accuracy': bal_acc,
        'ROC_AUC_OVR': roc_auc,
        'Confusion_Matrix': cm
    }

```

```python
# Regresión Logística Ridge
best_lr_ridge, best_params_lr_ridge, best_score_lr_ridge = run_grid_search(pipe_lr_ridge, param_grid_lr_ridge, df_features, y, cv)

# Regresión Logística Lasso
best_lr_lasso, best_params_lr_lasso, best_score_lr_lasso = run_grid_search(pipe_lr_lasso, param_grid_lr_lasso, df_features, y, cv)

# SVC con kernel RBF
best_svc_rbf, best_params_svc_rbf, best_score_svc_rbf = run_grid_search(pipe_svc_rbf, param_grid_svc_rbf, df_features, y, cv)

# SVC con kernel Poly
best_svc_poly, best_params_svc_poly, best_score_svc_poly = run_grid_search(pipe_svc_poly, param_grid_svc_poly, df_features, y, cv)

# SVC con kernel Sigmoid
best_svc_sigmoid, best_params_svc_sigmoid, best_score_svc_sigmoid = run_grid_search(pipe_svc_sigmoid, param_grid_svc_sigmoid, df_features, y, cv)

# MLP
best_mlp, best_params_mlp, best_score_mlp = run_grid_search(pipe_mlp, param_grid_mlp, df_features, y, cv)


models = {
    "LogisticRegression Ridge": best_lr_ridge,
    "LogisticRegression Lasso": best_lr_lasso,
    "SVC RBF": best_svc_rbf,
    "SVC Poly": best_svc_poly,
    "SVC Sigmoid": best_svc_sigmoid,
    "MLP": best_mlp
}

results = []
for model_name, model in models.items():
    model.fit(df_features, y)
    metrics = evaluate_model(model_name, model, df_features, y, cv)
    results.append(metrics)

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='Accuracy', ascending=False)

print("Ranking de modelos por Accuracy:")
print(results_df[['Model', 'Accuracy']])
```

# Bootstrap

```python
from sklearn.utils import resample
from sklearn.metrics import accuracy_score

model = best_lr_ridge  

n_bootstraps = 1000
boot_scores = []

for i in range(n_bootstraps):
    X_boot, y_boot = resample(df_features, y, replace=True, random_state=50 + i)
    model.fit(X_boot, y_boot)
    y_pred = model.predict(df_features)
    score = accuracy_score(y, y_pred)
    boot_scores.append(score)

boot_scores = np.array(boot_scores)
mean_acc = np.mean(boot_scores)
ci_lower = np.percentile(boot_scores, 2.5)
ci_upper = np.percentile(boot_scores, 97.5)

print("Bootstrap Accuracy Mean:", mean_acc)
print("95% Confidence Interval: [{:.3f}, {:.3f}]".format(ci_lower, ci_upper))

plt.figure(figsize=(8, 6))
plt.hist(boot_scores, bins=30, color='skyblue', edgecolor='k', alpha=0.7)
plt.axvline(mean_acc, color='red', linestyle='--', label=f"Mean = {mean_acc:.3f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"2.5th percentile = {ci_lower:.3f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"97.5th percentile = {ci_upper:.3f}")
plt.xlabel("Accuracy")
plt.ylabel("Frequency")
plt.title("Bootstrap Distribution of Accuracy")
plt.legend()
plt.show()
```

## Análisis de Bootstrap para la Métrica de Accuracy

En esta gráfica se presenta la **distribución del Accuracy** obtenida mediante **Bootstrap**. El procedimiento consiste en generar múltiples muestras aleatorias *con reemplazo* (en este caso, 1000) a partir del dataset original, entrenar el modelo en cada muestra y evaluar su rendimiento (accuracy) en el conjunto original. De esta forma se obtiene una **distribución empírica** de la métrica, a partir de la cual se puede estimar:

- **Media (Mean)**: La línea roja punteada indica el promedio del accuracy en las 1000 réplicas (aprox. 0.734).
- **Intervalo de Confianza (IC) al 95%**: Las líneas verdes punteadas señalan los percentiles 2.5 y 97.5, ofreciendo un rango donde es muy probable que se sitúe el accuracy real del modelo (aprox. [0.638, 0.810]).

### ¿Por qué aplicar Bootstrap?
1. **Variabilidad de la Métrica**: El Bootstrap permite conocer qué tan estable es el rendimiento del modelo. Al reentrenarlo en diferentes muestras (bootstraps), se observa si el accuracy varía significativamente.
2. **Intervalos de Confianza**: Brinda un rango más robusto sobre el cual reportar la métrica, en lugar de un único valor promedio.
3. **Tamaño de Muestra Limitado**: Cuando no se dispone de muchos datos, el Bootstrap ofrece una forma de aprovechar al máximo la información existente, reusando muestras con reemplazo.

### Interpretación de la Gráfica
- La **mayoría** de los valores de accuracy se concentran alrededor de 0.73-0.75, lo que indica que, en promedio, el modelo logra ese rendimiento.
- El **IC [0.638, 0.810]** sugiere que, dependiendo de la muestra, el accuracy podría caer tan bajo como ~0.64 o alcanzar hasta ~0.81.  
- Una distribución más estrecha (menos dispersa) significaría que el modelo es muy estable; una distribución muy ancha indicaría mayor sensibilidad a la muestra.

   - Añadir más datos o variables relevantes que ayuden a discriminar esas clases.

4. **Ajuste de Umbrales**: Si una clase es especialmente relevante (p.ej., "Déficit Alto"), podría modificar el umbral de decisión para incrementar la sensibilidad en esa clase.

```python
def bayesian_optimize(pipe, param_grid, X, y, cv, n_iter=30):
    """
    Optimización bayesiana simplificada usando GaussianProcessRegressor.
    Se parte de un grid inicial (definido en un diccionario param_grid) y se iteran n_iter veces
    para explorar nuevos candidatos.
    
    Devuelve:
      - best_params: Diccionario con el mejor conjunto de hiperparámetros.
      - best_score: Puntaje (accuracy) asociado al mejor conjunto.
    """
    keys = list(param_grid.keys())
    combos = list(itertools.product(*(param_grid[k] for k in keys)))
    
    X_params = []
    scores = []
    # Evaluar cada combinación inicial del grid
    for comb in combos:
        params_dict = dict(zip(keys, comb))
        pipe.set_params(**params_dict)
        cv_res = cross_validate(pipe, X, y, cv=cv, scoring='accuracy', n_jobs=1)
        score = np.mean(cv_res['test_score'])
        param_vector = []
        for k, val in params_dict.items():
            if isinstance(val, (int, float)):
                param_vector.append(float(val))
            else:
                param_vector.append(hash(val) % 997)
        X_params.append(param_vector)
        scores.append(score)
    
    X_params = np.array(X_params)
    scores = np.array(scores)
    
    # Escalar los vectores de hiperparámetros
    scaler_hp = StdScaler_hp()
    X_params_scaled = scaler_hp.fit_transform(X_params)
    
    kernel = 1.0 * RBF(length_scale=1.0)
    gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, random_state=42)
    gp.fit(X_params_scaled, scores)
    
    def propose_new_point(gp, X_params_scaled, scaler_hp):
        mins = X_params.min(axis=0)
        maxs = X_params.max(axis=0)
        random_points = []
        for _ in range(50):
            rp = [np.random.uniform(mn, mx) for mn, mx in zip(mins, maxs)]
            random_points.append(rp)
        random_points = np.array(random_points)
        random_points_scaled = scaler_hp.transform(random_points)
        mu, std = gp.predict(random_points_scaled, return_std=True)
        acquisition = mu + 1.96 * std  # Upper Confidence Bound
        best_idx = np.argmax(acquisition)
        return random_points[best_idx]
    
    for _ in range(n_iter):
        new_pt = propose_new_point(gp, X_params_scaled, scaler_hp)
        best_combo = None
        best_dist = float('inf')
        for comb in combos:
            cvec = []
            for k, val in dict(zip(keys, comb)).items():
                if isinstance(val, (int, float)):
                    cvec.append(float(val))
                else:
                    cvec.append(hash(val) % 997)
            cvec = np.array(cvec)
            dist = np.linalg.norm(cvec - new_pt)
            if dist < best_dist:
                best_dist = dist
                best_combo = comb
        params_dict = dict(zip(keys, best_combo))
        pipe.set_params(**params_dict)
        cv_res = cross_validate(pipe, X, y, cv=cv, scoring='accuracy', n_jobs=1)
        new_score = np.mean(cv_res['test_score'])
        new_vec = []
        for k, val in params_dict.items():
            if isinstance(val, (int, float)):
                new_vec.append(float(val))
            else:
                new_vec.append(hash(val) % 997)
        new_vec = np.array(new_vec)
        X_params = np.vstack([X_params, new_vec])
        scores = np.append(scores, new_score)
        X_params_scaled = scaler_hp.fit_transform(X_params)
        gp.fit(X_params_scaled, scores)
    
    best_idx = np.argmax(scores)
    best_params_vector = X_params[best_idx]
    best_score = scores[best_idx]
    best_combo = None
    best_dist = float('inf')
    for comb in combos:
        cvec = []
        for k, val in dict(zip(keys, comb)).items():
            if isinstance(val, (int, float)):
                cvec.append(float(val))
            else:
                cvec.append(hash(val) % 997)
        cvec = np.array(cvec)
        dist = np.linalg.norm(cvec - best_params_vector)
        if dist < best_dist:
            best_dist = dist
            best_combo = comb
    return dict(zip(keys, best_combo)), best_score


best_params_svc_rbf, best_score_svc_rbf = bayesian_optimize(pipe_svc_rbf, param_grid_svc_rbf, df_features, y, cv, n_iter=30)
pipe_svc_rbf.set_params(**best_params_svc_rbf)

best_params_svc_poly, best_score_svc_poly = bayesian_optimize(pipe_svc_poly, param_grid_svc_poly, df_features, y, cv, n_iter=30)
pipe_svc_poly.set_params(**best_params_svc_poly)

best_params_svc_sigmoid, best_score_svc_sigmoid = bayesian_optimize(pipe_svc_sigmoid, param_grid_svc_sigmoid, df_features, y, cv, n_iter=30)
pipe_svc_sigmoid.set_params(**best_params_svc_sigmoid)

best_params_mlp_bayes, best_score_mlp_bayes = bayesian_optimize(pipe_mlp, param_grid_mlp, df_features, y, cv, n_iter=30)
pipe_mlp.set_params(**best_params_mlp_bayes)

def evaluate_model(name, model, X, y, cv):
    y_pred = cross_val_predict(model, X, y, cv=cv, n_jobs=1)
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y, y_pred, average='weighted', zero_division=0)
    bal_acc = balanced_accuracy_score(y, y_pred)
    try:
        y_proba = cross_val_predict(model, X, y, cv=cv, n_jobs=1, method='predict_proba')
        classes = list(np.unique(y))
        y_bin = label_binarize(y, classes=classes)
        roc_auc = roc_auc_score(y_bin, y_proba, multi_class='ovr')
    except:
        roc_auc = np.nan
    cm = confusion_matrix(y, y_pred)
    return {
        'Model': name,
        'Accuracy': acc,
        'Precision_weighted': prec,
        'Recall_weighted': rec,
        'F1_weighted': f1,
        'Balanced_Accuracy': bal_acc,
        'ROC_AUC_OVR': roc_auc,
        'Confusion_Matrix': cm
    }

models = {
    "SVC RBF (Bayes)": pipe_svc_rbf,
    "SVC Poly (Bayes)": pipe_svc_poly,
    "SVC Sigmoid (Bayes)": pipe_svc_sigmoid,
    "MLP (Bayes)": pipe_mlp
}

results = []
for model_name, model in models.items():
    model.fit(df_features, y)
    metrics = evaluate_model(model_name, model, df_features, y, cv)
    results.append(metrics)

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='Accuracy', ascending=False)

print("Ranking de modelos por Accuracy:")
print(results_df[['Model', 'Accuracy']])
print("\nDetalles de todas las métricas:")
print(results_df)
```
